### 6.2 Categorical Features

#### 6.2.1 Trainig set
First let's take a look at the missing values in training set's categorical features: 

```python
print("Missing values in training set's categorical features:\n")
print(train_set.select_dtypes(exclude=[np.number]).isnull().sum().sort_values(ascending = False))
```

If we take a look at the descriptions of the categorical features that have missing values, we will see that having NA in almost all of them means the house doesn't have that feature. For example, NA in <i>Alley, Fence</i>, and <i>FireplaceQu</i> means the house has no alley access, fence, or fireplace respectively. The only feature that doesn't have NA as a category, is <i>MasVnrType</i>, it has "None" instead.
<i>MasVnrType</i> and <i>MasVnrArea</i> are related to each other, and having 0 <i>MasVnrArea</i> means the house <i>MasVnrType</i> is None. Since we replaced NAs in <i>MasVnrArea</i> to 0 in last steps, we are going to replace missing values of <i>MasVnrType</i> to None.

Finally, use one-hot encoding technique to make use of non-numeric information in modeling. One-hot encoding will transform categorical data into numbers so the model can understand whether or not a particular observation falls into one category or another.

## 10.Training And Running The Model
Note that since we are having a regression problem, we choose alinear activation function (ReLU) for the last layer.

```python
model= models.Sequential()
model.add(layers.Dense(512, activation='relu', input_shape = (train_set_X.shape[1], )))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(1))
```

(To check <b>keras.models.Sequential</b> documentation and examples click [here](https://keras.io/api/models/sequential/).)

```python
model.compile(optimizer = keras.optimizers.Adam(learning_rate=0.001),
              loss ='mae',
              metrics= ['mae', 'mse']
             )
```

(To check <b>keras's Model Training API</b> documentation and examples click [here](https://keras.io/api/models/model_training_apis/).)

```python
model.summary()
```

```python
model_fit = model.fit(train_set_X, train_set_Y,
                      epochs = 2000,
                      batch_size=5,
                      )
```

```python
plt.figure(figsize=(12, 12))
plt.plot(model_fit.history['mae'][20:], 'o', label='Training mae')
plt.title('Training mae')
plt.grid(True)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
plt.figure(figsize=(12, 12))
plt.plot(model_fit.history['mse'][20:], 'o', label='Training mse')
plt.title('Training mse')
plt.grid(True)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
train_set_Y_predictions = model.predict(train_set_X)
print ('R square accuracy:', r2_score(train_set_Y, train_set_Y_predictions))
print('Mean absolute error:', mean_absolute_error(train_set_Y, train_set_Y_predictions))
print ('Mean squared error:', mean_squared_error(train_set_Y, train_set_Y_predictions))
```
